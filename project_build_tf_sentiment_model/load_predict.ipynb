```python
import tensorflow as tf
```

```python
model = tf.keras.models.load_model('sentiment_model')
```

```python
from transformers import BertTokenizer
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-cased')
```

```python
### function that will prepare our data

def prep_data(text):
    tokens = tokenizer(text, max_length=512, truncation=True,
                      padding='max_length', add_special_tokens=True, return_tensors='tf')
    return {'input_ids': tokens['input_ids'], 'attention_mask': tokens['attention_mask']}
```

```python
prep_data('hello world')
```

```python
model.predict(prep_data('this movie was terrible'))
```

```python
probs = model.predict(prep_data('this movie was terrible'))[0]
probs
```

```python
import numpy as np
```

```python
np.argmax(probs)
```
